* 말뭉치를 사용할 경우 vocab을 어떻게 만들 것인가?

- character level(형태소 단위)
 - character 단위로 vocab을 만드는 방법
 - 한국어 기준으로 자음['ㄱ','ㄴ','ㄷ',…,'ㅎ'], 모음['ㅏ','ㅑ',…,'ㅣ'] 단위로 vocab을 나누거나 ['가','갸',…,'힣']와 같이 가능한 모든 글자를 vocab으로 나누는 것.
 - 가능한 모든 글자를 vocab으로 표현 가능하지만 각 단어의 고유한 의미를 표현하고 있는 것은 아니기 때문에 좋은 성능을 내지 못 하는 경우가 많음.

- space level(음절 단위)
 - 띄어쓰기 단위로 vocab을 만드는 방법
 - 띄어쓰기로 할 경우 한걱어의 경우는 조사/어미 등으로 인해서 중복단어 문제가 발생
 - ex: 책 - ['책이','책에','책을',...]
 - vocab이 매우 커지게 되고 빈도수가 낮은 단어들은 잘 학습이 되지 않음.
 - 대안으로 vocab을 줄이기 위해서 일정 빈도 이상인 단어만 vocab으로 만들지만, 이 경우에 없는 단어는 unknown으로 처리해야하는 문제 발생

- subword level
 - 많은 단어를 처리하면서도 unknown이 발생할 확률을 줄이는 방법
 - 단어의 빈도 수를 계산해서 subword 단위로 쪼개는 방법
 - google sentensepiece 활용
 - subword(BPE) 참고: https://wikidocs.net/22592
